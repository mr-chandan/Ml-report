Importing the Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load dataset into a DataFrame

```python
data = pd.read_csv('datasets/rice_production_by_country.csv')
```

```python
data.head()
```

Bar graph

```python
data_sorted = data.sort_values(by='Rice Production (Tons)', ascending=False)

plt.figure(figsize=(50,40))  # Increase the figure size for better visualization
sns.barplot(x='Country', y='Rice Production (Tons)', data=data_sorted, order=data_sorted['Country'], orient='v')
plt.title('Rice Production (Tons) by Country', fontsize=40)  # Increase title font size
plt.xticks(rotation=45, ha='right', fontsize=20)  # Rotate x-axDis labels further, align right, and increase font size
plt.yticks(fontsize=20)  # Increase y-axis labels font size
plt.xlabel('Country', fontsize=30)  # Increase x-axis label font size
plt.ylabel('Rice Production (Tons)', fontsize=30)  # Increase y-axis label font size
plt.tight_layout()  # Adjust layout to prevent overlapping elements
plt.ylim(reversed(plt.ylim()))
plt.show()
```

Histogram

```python
rice_production_per_person = data['Rice Production Per Person (Kg)']

plt.figure(figsize=(50,40))  
plt.hist(rice_production_per_person, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Rice Production Per Person', fontsize=30)  
plt.xlabel('Rice Production Per Person (Kg)', fontsize=25)  
plt.ylabel('Frequency', fontsize=25) 
plt.grid(True) 
plt.xticks(rotation=45, ha='right', fontsize=20)
plt.yticks(fontsize=20)  
plt.tight_layout() 
plt.show()
```

Heatmap

```python
numeric_data = data.drop(columns=['Country'])

for col in numeric_data.columns:
    numeric_data[col] = numeric_data[col].replace(r'[^\d.]', '', regex=True).astype(float)

correlation_matrix = numeric_data.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5, annot_kws={"size": 12})
plt.title('Correlation Heatmap of Rice Production Features', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

Scatter plot

```python
plt.figure(figsize=(50, 40)) 
sns.scatterplot(x='Rice Production (Tons)', y='Rice Acreage (Hectare)', data=data)
plt.title('Rice Production (Tons)', fontsize=30)  
plt.xlabel('Rice Acreage (Hectare)', fontsize=25)  
plt.ylabel('Rice Yield (Kg / Hectare)', fontsize=25)  
plt.xticks(rotation=45, ha='right', fontsize=15)
plt.yticks(fontsize=15) 
plt.show()
```

Pairplot

```python
sns.pairplot(data, hue='Rice Production (Tons)', diag_kind='kde', height=5)
plt.show()
```
